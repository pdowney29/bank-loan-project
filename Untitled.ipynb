# Bank Loan Project

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, roc_curve, roc_auc_score, classification_report, log_loss, make_scorer
import seaborn as sn
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV, train_test_split
import sklearn
from sklearn.svm import SVR
```

# EDA

```python
df = pd.read_csv('loan.csv')
df.head()
```

```python
df.describe().T
```

```python
df.info()
```

```python
df.hist(figsize=(8,12),bins=3);
```

```python
df.corr()
```

```python
pd.crosstab(df['Loan.Length'], df['Loan.Amount'], rownames=['Length'])
```

```python
df['FICO.Score'].fillna(df['FICO.Score'].mean(), inplace = True)
```

```python
df.dropna(subset=['Interest.Rate'])  #drop rows where no interest rate
```

```python
df.dropna(inplace=True)
df
```

```python
df['Interest.Rate'] = (df['Interest.Rate'] < 9.0).astype(int)
```

```python
df.head()
```

```python
df.describe().T
```

```python
# class imbalance
df['Interest.Rate'].sum() / 2493, (df['Interest.Rate'].size - df['Interest.Rate'].sum()) / 2493
```

# Split and Scale

```python
y = df.pop('Interest.Rate')
X = df
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=21)
# If not scaling ...
# Convergence Warning: The max_iter was reached which means the coef_ did not converge
scalar = StandardScaler().fit(X_train)
X_train = scalar.transform(X_train)
X_test = scalar.transform(X_test)
```

```python
y_train
```

```python
print('\n'.join(sklearn.metrics.SCORERS.keys()))
```

# Gridsearch¶

```python
params = {
          'penalty': ['l2'], 
          'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
          'fit_intercept': [True, False],
          'C': np.logspace(-3,3)
         }
gs = GridSearchCV(LogisticRegression(), param_grid=params, scoring='precision', cv=10, n_jobs=5).fit(X_train, y_train)
gs
```


# Results

```python
gs.best_estimator_
```

# Final Model

```python
gs.best_params_
```

```python
model = LogisticRegression(**gs.best_params_).fit(X_train, y_train)
```

```python
y_hat = model.predict(X_test)
p_hat = model.predict_proba(X_test)[:,1]
```

```python
log_loss(y_test, p_hat)
```

```python
baseline = 1 - (y_test.sum() / y_test.size)
accuracy = accuracy_score(y_test, y_hat)
baseline, accuracy
```

```python

# TP / (TP + FP) # type 1 errors
prec = precision_score(y_test, y_hat)
```

```python

# TP / (TP + FN) # type 2 errors ---
recall = recall_score(y_test, y_hat)
```

# ROC Curve¶

```python
auc = roc_auc_score(y_test, p_hat)
auc
```

```python
fpr, tpr, thr = roc_curve(y_test, p_hat)
for th, fp, tp in zip(thr, fpr, tpr):
    print("{0:.2f} {1:.2f} {2:.2f}".format(th, fp, tp))
```

```python
plt.plot(fpr, tpr, label=auc, color='#ff33cc', alpha=0.7)
plt.plot(np.linspace(0,1), np.linspace(0,1), linestyle='--', alpha=0.5)
plt.title('ROC Curve')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
plt.show()
print('accuracy', accuracy, 'precision', prec, 'recall', recall)


```
